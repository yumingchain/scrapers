```python
from playwright.async_api import async_playwright

import pandas as pd
```

```python

pw = await async_playwright().start()
browser = await pw.chromium.launch(headless=T)
context = await browser.new_context()
page = await context.new_page()
```

```python
url = "https://ujsportal.pacourts.us/CaseSearch"
await page.goto(url, timeout=120000)
```

```python
search_by = "Date Filed"
await page.locator('select[title="Search By"]').select_option(
            label=search_by
        )
```

```python
start_date="2024-03-25"
end_date="2024-03-28"
await page.fill('input[name="FiledStartDate"]',end_date )
await page.fill('input[name="FiledEndDate"]', end_date)
```

```python
await page.locator("#btnSearch").click()
await page.wait_for_timeout(7000)
```

```python
# id="caseSearchResultGrid"
#get the table
table = await page.query_selector("#caseSearchResultGrid")
```

```python
#get all the rows
rows = await table.query_selector_all("tr")
```

```python
await rows[0].inner_text()
```

```python
header_table = await table.query_selector("thead > tr")
```

```python
columns_elements = await header_table.query_selector_all("th")
columns_elements
```

```python
columns = [await column.inner_text() for column in columns_elements]
columns
```

```python
#get all the cells
cells = await rows[2].query_selector_all("td")
cells
```

```python
items = [await cell.inner_text() for cell in cells]
items
```

```python
await cells[-1].eval_on_selector_all("a.icon-wrapper", "elements => elements.map(element => element.href)")
```

```python
#save file pdf
hrefs = await page.eval_on_selector_all("a.icon-wrapper", "elements => elements.map(element => element.href)")
hrefs
```

```python
#zip the columns and the items
data = dict(zip(columns, items))
data
```

```python
#loop through the rows
data = []
for row in rows:
    cells = await row.query_selector_all("td")
    items = [await cell.inner_text() for cell in cells]
    items_dict =  dict(zip(columns, items))
    data.append(items_dict)
```

```python
data
```

```python
#to csv
df = pd.DataFrame(data)
```

```python
df.shape, df.describe(), df.columns
```

```python
df.to_csv("data.csv", index=False)
```
